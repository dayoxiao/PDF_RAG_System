# By llamaIndex itself

```python
documents = SimpleDirectoryReader("./test_doc").load_data()

# embedding model
Settings.embed_model = OllamaEmbedding(model_name="bge-m3:latest")

# ollama llm
Settings.llm = Ollama(model="deepseek-r1:7b", request_timeout=360.0)
#Settings.llm = Ollama(model="qwen2.5:latest", request_timeout=360.0)

index = VectorStoreIndex.from_documents(
    documents,
)
```

```python
# data query
query_engine = index.as_query_engine()
response = query_engine.query("我是部長，要出差，費用應該如何處理？")
print(response)
#print("\n########## Below are retrieved data ##########")
display([(n.text, n.metadata) for n in response.source_nodes])
```

```python
response = query_engine.query("我是科主管，要在國內出差，可以申請多少全日雜費、宿費、交通費？")
print(response)
#print("\n########## Below are retrieved data ##########")
display([(n.text, n.metadata) for n in response.source_nodes])
```

```python
# chat
memory = ChatMemoryBuffer.from_defaults(token_limit=1500)

chat_engine = index.as_chat_engine(
    chat_mode="context",
    memory=memory,
    system_prompt=(
        "You are a chatbot assistance, help and answer user's quetion precisely from the document."
    ),
)

response = chat_engine.chat("我是部長，要出差，費用應該如何處理？")
print(response)
```

```python
chat_engine.reset()
```

# By llamaindex and docling, with markdown export

```python
# PyPdfium without EasyOCR
# --------------------
pipeline_options = PdfPipelineOptions()
pipeline_options.do_ocr = False
pipeline_options.do_table_structure = True
pipeline_options.table_structure_options.do_cell_matching = False
#pipeline_options.table_structure_options.do_cell_matching = True

pipeline_options.images_scale = 2
pipeline_options.generate_page_images = True
pipeline_options.generate_picture_images = True

pypdfium_converter = DocumentConverter(
    format_options={
        InputFormat.PDF: PdfFormatOption(
        pipeline_options=pipeline_options, backend=PyPdfiumDocumentBackend
        )
    }
)
```

```python
reader = DoclingReader(doc_converter=pypdfium_converter)
node_parser = MarkdownNodeParser()
embed_dim = len(Settings.embed_model.get_text_embedding("你好"))

dir_reader = SimpleDirectoryReader(
    input_dir='./test_doc',
    file_extractor={".pdf": reader},
)

vector_store_md = MilvusVectorStore(
    uri=str(Path(mkdtemp()) / "docling_md.db"),  # or set as needed
    dim=embed_dim,
    overwrite=True,
)

index = VectorStoreIndex.from_documents(
    documents=dir_reader.load_data(),
    transformations=[node_parser],
    storage_context=StorageContext.from_defaults(vector_store=vector_store_md),
    embed_model=Settings.embed_model,
)
result = index.as_query_engine(llm=Settings.llm).query("我是部長，要出差，費用應該如何處理？")
print(result)
display([(n.text, n.metadata) for n in result.source_nodes])
```

```python
result = index.as_query_engine(llm=Settings.llm).query("我是科主管，要在國內出差，可以申請多少全日雜費、宿費、交通費？")
print(result)
display([(n.text, n.metadata) for n in result.source_nodes])
```

```python
# chat
memory = ChatMemoryBuffer.from_defaults(token_limit=1500)

chat_engine = index.as_chat_engine(
    chat_mode="context",
    memory=memory,
    system_prompt=(
        "你是一個協助回答公司相關業務的聊天機器人，請根據提供的文檔精確幫助用戶獲得解答"
    ),
)

response = chat_engine.chat("我是部長，要出差，費用應該如何處理？")
print(response)
display([(n.text, n.metadata) for n in response.source_nodes])
```

```python
response = chat_engine.chat("我是科主管，要國內出差，可以申請多少全日雜費、宿費、交通費？")
print(response)
display([(n.text, n.metadata) for n in response.source_nodes])
```

```python
response = chat_engine.chat("在國內出差要點中，「國內出差雜、宿、交通費標準」裡，其他人員的全日雜費可以申請多少元？")
print(response)
display([(n.text, n.metadata) for n in response.source_nodes])
```

```python
response = chat_engine.chat("給我國內出差要點中，附件：國內出差雜、宿、交通費標準的全部內容")
print(response)
display([(n.text, n.metadata) for n in response.source_nodes])
```

```python
chat_engine.reset()
```
